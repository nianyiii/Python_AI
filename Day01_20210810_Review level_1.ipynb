```python
# list -> for, "while", list comprehension 產生 10 個亂數 [1, 6]

scores = []
times = 10

while True:
    if times <= 0:
        break
    scores.append(random.randint(1, 6))
    times = times - 1
    print('times', times)
scores
```

```python
# list -> for, "while", list comprehension 產生 10 個亂數 [1, 6]

import random

number = []
n = 1

while n <= 10:
    n += 1
    number.append(random.randint(1, 6))
number
```

```python
# list -> for, "while", list comprehension 產生 10 個亂數 [1, 6]

import random

score = []
x = 10

while (x > 0) :
    score.append(random.randint(1, 6))
    x -= 1    
score
```

```python
%%time

# 題目：產生1K個1-6數字，請畫圖找出每一個數字數量（畫出Bar圖）

import random

times = 1000
scores = [random.randint(1, 6) for i in range(times)]
scores
# 執行時間快

# list -- dict  數字1假設有四個，以此類推
# distribution = {1:4, 2:3, 3:5, 4:8, 5:2, 6:1}
```

```python
# 非大數據語言（JAVA NET PHP）寫法

scores = [random.randint(1, 6) for i in range(times)]
s1, s2, s3, s4, s5, s6 = 0, 0, 0, 0, 0, 0

for s in scores:
    if s == 1:
        s1 = s1 + 1
    if s == 2:
        s2 = s2 + 1
    if s == 3:
        s3 = s3 + 1
    if s == 4:
        s4 = s4 + 1
    if s == 5:
        s5 = s5 + 1
    if s == 6:
        s6 = s6 + 1
print(s1, s2, s3, s4, s5, s6)
```

```python
# Pandas寫法

import pandas as pd

scores = [random.randint(1, 6) for i in range(times)]

pd.Series(scores).value_counts().sort_index()
```

```python
# Pandas寫法（圖表呈現）

import pandas as pd

scores = [random.randint(1, 6) for i in range(times)]

pd.Series(scores).value_counts().sort_index().plot(kind = 'bar', figsize = (16,4))
```
