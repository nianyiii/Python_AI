### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import requests #讀出內容無規則性，資料全部串接無法看清楚段落分行

r = requests.get('https://api.github.com/events')
r.text
```

```python
import json #讀出內容清楚，條列式呈現，較易閱讀

r = requests.get('https://api.github.com/events')

json.loads(r.text)
```

### 自行尋找一個合適的 API 接口做練習，並且查看其回傳內容

* https://cat-fact.herokuapp.com/facts (來源：https://alexwohlbruck.github.io/cat-facts/)
* http://odata.wra.gov.tw/v4/RealtimeWaterLevel (來源：https://data.gov.tw/dataset/25768)


```python
payload = {'key1':'value1','key2':'value2'}
r = requests.get('https://cat-fact.herokuapp.com/facts', params = payload)
json.loads(r.text)

#payload定義：即承載量，在一堆資料中我們所“關心”的部分。
#網路定義：On the Internet, a payload is either: The essential data is being carried within packet or other transmission unit.
#資料：http://noelsaga.herokuapp.com/blog/2015/07/18/kai-fa-zhong-,chang-jian-de-can-shu-payloadshi-shi-mo
```

```python
r = requests.post('https://cat-fact.herokuapp.com/facts', data = {'key':'value'})
json.loads(r.text)
```
