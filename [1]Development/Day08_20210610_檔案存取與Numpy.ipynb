```python
# Practice

no = 1
scores = dict()

while True:
    score = int(input('請輸入第 '+ str(no)+ ' 號的成績：(-1結束)' ))
    if score == -1: break;
    scores[no] = score  # 將取得的分數根據第幾次轉換成dict
    no += 1

# 將dict轉換成str寫入文字檔
file = open('score.txt', 'w', encoding = 'utf-8')
file.write(str(scores))
```

```python
# Practice

f = open('data1.txt', 'r', encoding = 'utf-8')
result = list()
for line in f.readlines():
    line = line.strip()
    if not len(line) or line.startswith('#'): # 讀取資料列為空白或開頭為#就捨去，亦可寫成if len(line)=0。
        continue
    result.append(line)
result.sort()
print(result)
open('result-readlines.txt', 'w', encoding = 'utf-8').write('%s' % '\n'.join(result)) # 也可省略['%s' %]
# 將讀取後結果去除空白列或開頭為#的資料列，再寫出至新檔案
```

#### json操作

```python
# json資料操作
# dump(s)：將數據轉換成json格式檔案（變數）

import json

# dict變數
dict1 = {'Python':'好',
        'gjun':100,
       'Python-class':True,
       'ICQ':None}

# json變數
json2 = json.dumps(dict1, ensure_ascii = False) # 執行dumps時，加入ensure_ascii = False參數可避免亂碼產生
print('json變數:', json2)

json3 = json2.encode('utf-8')
print('json變數utf-8:', json3)
```

```python
# Practice

import json

# dict變數
dict1 = {'Python':'好',
        'gjun':100,
       'Python-class':True,
       'ICQ':None}

# json變數
json2 = json.dumps(dict1, ensure_ascii = False)

# 開啟檔案，輸出dict1結果
with open('data0.json', 'w') as outfile:
    json.dump(dict1, outfile)

# 開啟檔案，輸出json變數結果
file = open('data1.json', 'w', encoding = 'utf-8')
file.write(json2)
file.close()
```

```python
# json資料操作
# load(s)：從json字串的檔案(記憶體）中取出資料轉入Python的dict資料型態

import json

message = '{"python":"good", "gjun":100, "Python-class":true, "ICQ":null}'
json2 = json.loads(message) # 將json字串轉換成dict
print(json2, type(json2))
```

## Numpy

```python
# Numpy：陣列類似list，但當中所有元素必須是相同資料型態，如int、float、string
# List：每一個元素可以不同資料型態，如int、float、string
# List vs Numpy

# List
scores = [12, 46, 33, 68, 51]
(scores, type(scores))
```

```python
# Numpy
import numpy as np
np.array(scores), type(np.array(scores))
```

```python
# 將以上列表中的分數再加上30分
# List無法“直接”使用算術運算工具（+, -, *, /）

# List
# scores + 40 ---> 無法直接運算
[s + 40 for s in scores]
```

```python
# 承上，Numpy提供更有效率計算與多維運算工具（批次運算）

# Numpy
import numpy as np
np.array(scores) + 40
```
