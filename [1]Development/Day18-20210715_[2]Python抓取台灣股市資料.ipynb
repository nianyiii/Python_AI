```python
# 匯入套件並且宣告變數進行網路資料抓取

import requests 
import pandas as pd
datestr = '20210715'
r = requests.post('http://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date=' + datestr + '&type=ALL')
```

```python
# 針對抓取完成的資料進行分解，過程中使用\r\n進行分行

tickers = [] #宣告List
for i in r.text.split('\r\n'):
    if (len(i.split('",')) == 17 and i[0] != '='):
        _ticker = i.split('",') #使用",區隔每一個價格
        _ticker = [t.replace('"','') for t in _ticker] #將左邊的" 取代成空字元
        _ticker = [t.replace(',','') for t in _ticker] #將數字千分為 取代為空字元
        tickers.append(_ticker)
```

```python
# 查看抓取的資料
r.text
```

```python
# 檢查股價名稱與找出第一筆資料

print(tickers[0:5])
```

```python
# 轉換成資料集格式稱之為pandas.DataFrame

df = pd.DataFrame(tickers)
df
```

```python
#將第一列的資料換成標題，並且刪除第一列

df.columns = df.loc[0, :]
df.drop(0, axis = 0 , inplace = True)
df
```

```python
# According to datestr then output df as datestr.csv
# For example 20210715.csv -> c:\Python

df.to_csv('c:\Python\datestr.csv')
```
