## 陣列資料增減

```python
# np.append()：在陣列末端附加資料。
# numpy附加（注意軸的問題：axis = None, 轉成1-rank ; axis = 0, 上下 ; axis = 1, 左右）

import numpy as np

a = np.array([[1,2,3], 
              [4,5,6]])
print(a)
print()

b = np.append(a, [7,8,9]) # 沒寫入axis等同None
print(b)
```

```python
# 承上範例，axis=0的操作

import numpy as np

a = np.array([[1,2,3], 
              [4,5,6]])
print(a)
print()

b = np.append(a, [[7,8,9]], axis = 0) # a為2-rank, 因此附加的資料也要2-rank
print(b)
```

```python
# 承上範例，axis=1的操作

import numpy as np

a = np.array([[1,2,3], 
              [4,5,6],
              [8,8,8]])
print(a)
print()

b = np.append(a, [[7],
                  [8],
                  [9]], axis = 1)
print(b)
```

#### 範圍與百分位數

```python
# 讀取diabetes.csv

import pandas as pd

check = pd.read_csv('diabetes.csv')
age = check['Age'].values
age
```

```python
# 承上
# np.ptp()：可取指定軸的值範圍（最大 - 最小）
# np.percentile()：代表百分位數，用於統計數據度量指標。參數：（a = array, q = percent, axis）

import numpy as np
import matplotlib.pyplot as plt

print(np.ptp(age), age.min(), age.max())

plt.boxplot(age)
plt.show()

print(np.percentile(a = age, q = 50))
```

```python
# Pracitce

import numpy as np

a = np.array([[0, 30,45],
              [60,75,90]])
print(a)
print()
b = np.ptp(a, axis = 0)
print(b)
print()
b = np.ptp(a, axis = 1)
print(b)
```

```python
# Practice

import numpy as np

a = np.array([[30,40,70],
              [80,20,10],
              [50,90,60]])
print(a)
print()
print(np.sort(a.ravel())) # 攤平後，再排序
print()
b = np.percentile(a, 50) # 取中位數
print(b)
print()
b = np.percentile(a, 50, axis = 0)
print(b)
print()
b = np.percentile(a, 50, axis = 1)
print(b)
print()
```

#### 變異數與標準差

```python
# np.var()：由陣列取得變異數的方法 = 變異數：(xi - mean(x)) ** 2 .sum() / records個數

# 讀取diabetes.csv
import pandas as pd

check = pd.read_csv('diabetes.csv')
age = check['Age'].values

print(((age - np.mean(age)) ** 2 ).sum() / len(age))

np.var(age)
```

```python
# 承上
# np.std()：由陣列取得標準差的方法 = 標準差：var ** 0.5

print((((age - np.mean(age)) ** 2 ).sum() / len(age)) ** 0.5)

np.std(age)
```

#### 累加與累差

```python
# np.cumsum()：進行指定軸資料的累加。參數：a = array, axis，若未指定軸，則攤平後進行累加。
# 累加：維持原有資料數量，在計算結果上進行下一個計算。

import numpy as np

a = np.array([1, 2, 3, 4])
np.cumsum(a)
```

```python
# np.diff()：進行指定軸資料的累差。參數：（a = array, axis），若未指定軸，預設axis = 1。
# 累差：以原資料進行資料內容刪除，若axis為0則減少一個row，若axis為1則減少一個column。

import numpy as np

a = np.array([1, 2, 3, 4])
np.diff(a)
```

```python
# Practice

import numpy as np

a = np.array([[1, 2, 3], 
              [13, 6, 9], 
              [12, 24, 36]])
print(a.ravel())
print("---------------")
print(np.cumsum(a))
print("---------------")
print(np.cumsum(a, axis = 0))
print("---------------")
print(np.cumsum(a, axis = 1))
```
