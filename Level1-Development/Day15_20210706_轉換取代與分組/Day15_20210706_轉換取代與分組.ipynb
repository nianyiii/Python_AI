```python
import pandas as pd

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'KINGS', 'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
            'Rank': [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
            'Year': [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
            'Points':[876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690]}
df = pd.DataFrame(ipl_data)
df
```

```python
# .capitalize()：回傳將str改成首字母大寫，其餘字母小寫。
'kings'.capitalize(), 'KINGS'.capitalize()
```

```python
# 讓Team欄位資料一致
df['Team'] = df['Team'].apply(lambda x:x.capitalize())
df
```

```python
# 根據指定欄位排序（依照年份）
df.sort_values(by = ['Year', 'Team'])
```

```python
# 分群
for name, data in df.groupby(by = 'Year'):
    print(name, '\n', data, '\n')
```

```python
# 轉成dict方法1
for name, data in df.groupby(by = 'Year'):
    print({name:data})
```

```python
# 轉成dict方法2（同方法1）
for g in df.groupby(by = 'Year'):
    print({g[0]:g[1]})
```

```python
# 亦可先轉成list
list(df.groupby(by = 'Year'))
```

```python
# 承上，再轉成dict
dict(list(df.groupby(by = 'Year')))
```

```python
import pandas as pd
import numpy as np

df3 = pd.DataFrame(np.random.randint(1, 10, (5,5)), 
                   columns = ['a', 'b', 'c', 'd', 'e'], 
                   index = ['A', 'B', 'C', 'D', 'E'])
df3.iloc[1, 1:4] = np.NaN
df3
```

```python
# 將欄位改成對應電腦語言名稱
mapping_dict = {'a':'Python', 'b':'Python', 'c':'Java', 'd':'C', 'e':'Java'}
df4 = df3
df4.rename(columns = mapping_dict)
```

```python
print("size")
df3.groupby(mapping_dict, axis = 1).size() # 欄位個數
```

```python
print("count")
df3.groupby(mapping_dict, axis = 1).count() # 資料筆數
```

```python
print("sum")
df3.groupby(mapping_dict, axis = 1).sum() # 根據相同欄位名稱進行加總
```

```python
import pandas as pd
import numpy as np

a = pd.MultiIndex.from_arrays(
           [['Python', 'Java', 'Python', 'Java', 'Python'], 
            [ 'A', 'A', 'B', 'C', 'B']],
    names = ['language', 'index'])
df4 = pd.DataFrame(np.random.randint(1, 10, (5, 5)), columns = a)
df4
```

```python
# 以language進行分組
df4.groupby(level = 'language', axis = 1).sum()
```

```python
# 以index進行分組
df4.groupby(level = 'index', axis = 1).sum()
```

```python
import pandas as pd

sales = pd.read_csv('salesv3.csv', encoding = 'utf-8')
sales
```

```python
# 查看date的型態
sales.dtypes
```

```python
# 將資料date欄位轉成datetime型態
sales['date'] = pd.to_datetime(sales['date'])
```

```python
# 再次查看date的型態
sales.dtypes
```

```python
# 改變資料集型態
sales.set_index('date')
```

```python
# .head()：取出前幾筆資料（DataFrame用法）
sales.set_index('date').head(10) # 前10筆，或寫成sales.set_index('date')[:10]
```

```python
# 根據索引資料，按照月份（至月底），且針對指定欄位進行資料筆數or數值加總
sales.set_index('date').resample('M')['ext price'].size() # M月份；D日
```

```python
# 根據產品名稱，按照每日，針對指定欄位進行資料筆數or數值加總（有索引的因素）
sales.set_index('date').groupby('name')['ext price'].resample('D').sum()
```

```python
# 將欄位date改名稱（date屬於專有名稱，較不適合直接使用）

import pandas as pd

sales = pd.read_csv('salesv3.csv', encoding = 'utf-8')
sales = sales.rename(columns = {'date':'orderdate'})
sales.head(10)
```

```python
# 查看orderdate欄位型態
sales.dtypes
```

```python
# 將資料orderdate欄位轉成datetime型態
sales['orderdate'] = pd.to_datetime(sales['orderdate'])
```

```python
# 再次查看orderdate欄位型態
sales.dtypes
```

```python
# 根據name欄位分群，計算ext price欄位的筆數or數值加總
sales.groupby('name')['ext price'].sum() # 型態為Series
```

```python
# 承上，將ext price多一個中括號，結果將與下方相同
sales.groupby('name')[['ext price']].sum() # 型態為DateFrame
```

```python
# 根據name、ext price兩個欄位，再從name欄位進行筆數or數值加總
sales[['name', 'ext price']].groupby('name').sum() # 型態為DataFrame
```

```python
# 補充：從Series資料型態中，針對日期時間欄位取出年or月or日
sales['orderdate'].dt.day
```

```python
# 補充：從DataFrame資料型態，可直接取出年or月or日
pd.to_datetime('20210709').month
```

```python
# pd.Grouper：用法與groupby相同，將資料進行分組
sales.groupby(['name', pd.Grouper(key = 'orderdate', freq = 'M')])['ext price'].sum()
# 與sales.set_index('orderdate').groupby('name')['ext price'].resample('M').sum() 結果相同
```

```python
# 承上，將ext price欄位再加上一個中括號，資料型態為DataFrame
sales.groupby(['name', pd.Grouper(key = 'orderdate', freq = 'M')])[['ext price']].sum()
```

```python
# 將freq設定為A-DEC，取出資料至12月底（可更換其他月份）
sales.groupby(['name', pd.Grouper(key = 'orderdate', freq = 'A-DEC')])[['ext price']].sum()
```

```python
# 簡易匯整，直接進行加總（不分組）
sales[['ext price', 'quantity']].sum()
```

```python
# 設定by參數時，如果將未指定的欄位放進去，則產生錯誤，應改為下方寫法。
sales[[ext price', 'quantity']].groupby(by = 'name').sum()
```

```python
# 針對指定欄位-->分組-->進行加總
sales[['name', 'ext price', 'quantity']].groupby(by = 'name').sum()
```

```python
# 分組-->針對指定欄位-->進行加總
sales.groupby(by = 'name')[['ext price', 'quantity']].sum()
```

```python
# 簡易彙整，直接進行平均（不分組）
print(sales['unit price'].mean())
print(type(sales['unit price'].mean()))
```

```python
# 承上，多一個中括號會顯示欄位名稱
print(sales[['unit price']].mean())
print(type(sales[['unit price']].mean()))
```

```python
# .agg()：整併各種計算為多個column輸出。
# 如果筆數和數值的加總需要一起呈現
sales.groupby(by = 'name')[['ext price', 'quantity']].agg(['size', 'sum'])
```
