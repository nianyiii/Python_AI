```python
import speech_recognition as sr
r = sr.Recognizer()
```

```python
mic = sr.Microphone()
```

```python
sr.Microphone.list_microphone_names()
```

```python
# 設置麥克風作為音訊來源
with mic as source:
    print('說些什麼吧！')
    audio = r.listen(source)
```

```python
# 將語音轉為文字
try:
    text = r.recognize_google(audio, language = 'zh-TW')
    print('您說的是：' + text)
except sr.UnknownValueError:
    print('語音無法識別！')
    text = '語音無法識別'
except sr.RequestError as e:
    print('無法連接到語音識別服務：{0}'.format(e))
    text = '無法連接到語音識別服務'
```

#### Conversation

```python
import speech_recognition

def listenTo():
    r = speech_recognition.Recognizer()

    with speech_recognition.Microphone() as source:
        r.adjust_for_ambient_noise(source)
        audio = r.listen(source)

    return r.recognize_google(audio, language = 'zh-TW')
```

```python
import tempfile
from gtts import gTTS
from pygame import mixer
mixer.init()

def speak(sentence):
    with tempfile.NamedTemporaryFile(delete = True) as fp:
        tts = gTTS(text = sentence, lang = 'zh-tw')
        tts.save("{}.mp3".format(fp.name))
        mixer.music.load('{}.mp3'.format(fp.name))
        mixer.music.play()
        
speak('大家好!祝大家有個美好的一天')
```

```python
# 驗證聽我所說，再回答
speak(listenTo())
```

```python
Answer = {
  '你好嗎' : '我很好',
  '你很帥' : '謝謝啦 !',
  '再見了' : '下次見了! 再聊! 拜拜'
}
```

```python
speak(Answer.get(listenTo(), '對不起，聽不清楚! 請再說一變，謝謝 ! 再回答你'))
```

```python
import librosa
songname = 'violin.wav'
```

```python
import matplotlib.pyplot as plt

cmap = plt.get_cmap('Greys_r') # Greys_r inferno
plt.figure(figsize = (10,10))

y, sr = librosa.load(songname, mono = True, duration = 5)
plt.specgram(y, NFFT = 2048, Fs = 2, Fc = 0, noverlap = 128, cmap = cmap, 
             sides = 'default', mode = 'default', scale = 'dB');
plt.axis('off');
plt.savefig('./violin.wav.png')
plt.clf()
```

```python
# 將每一種歌曲型態轉換成Spectrogram
# 參考CNN MODEL進行分類
img = plt.imread('./violin.wav.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```
