```python
# 畫出"Temperature" Hist

import pandas as pd
import matplotlib.pyplot as plt

lemon = pd.read_csv('lemonade.csv')
lemon['Temperature'].plot(kind = 'hist', edgecolor = 'red', bins = 12, figsize = (12, 4))
plt.show()
print(lemon['Temperature'].describe())
```

```python
# 畫出"Rainfall"與"Sales" Scatter

import matplotlib.pyplot as plt

plt.scatter(lemon['Rainfall'], lemon['Sales'])
```

```python
# 畫出"Rainfall"與"Sales"的對數 Scatter

import matplotlib.pyplot as plt
import numpy as np

plt.scatter(np.log(lemon['Rainfall']), np.log(lemon['Sales']))
```
