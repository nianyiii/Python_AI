```python
# 建議不要使用np.matrix改使用np.array
# 矩陣乘法：深度學習CNN是利用矩陣改變圖片
# https://en.wikipedia.org/wiki/Kernel_(image_processing)

import numpy as np

A = np.matrix([[1, 2], [4, 5]])
B = np.matrix([[9, 2], [7, 6]])

A + B, A - B, A==B, A * B , A @ B # A*B = A@B
```

```python
import numpy as np

A = np.array([[1, 2], [4, 5]])
B = np.array([[9, 2], [7, 6]])

# array使用乘法，表示元素*元素 
A + B, A - B, A==B, A * B , np.dot(A, B), A.dot(B) # np.dot(A, B) = A.dot(B)
```

```python
# Practice：矩陣相乘

import numpy as np

A = np.array([[1, 2, 3],
              [4, 5, 6]])
B = np.array([[9, 8],
              [7, 6],
              [5, 4]])

print(np.dot(A, B))
print(A.dot(B))
print(A @ B)
```

```python
# 母體（Population）標準差

import pandas as pd
import numpy as np

data = pd.read_csv('Lemonade.csv')
sales = data['Sales'].values
mu = np.sum(sales) / len(sales)
print('average', mu)
var = np.sum((sales - mu) ** 2) / len(sales)
print('variance', var)
std = var ** 0.5
print('std', std)
```

```python
# .describe()：Generate descriptive statistics

import pandas as pd

data = pd.read_csv('Lemonade.csv')
data[['Sales']].describe() # -> pandas standard deviation
```

```python
# 樣本（Sample）標準差

import pandas as pd
import numpy as np

data = pd.read_csv('Lemonade.csv')
sales = data['Sales'].values
mu = np.sum(sales) / len(sales)
print('average', mu)
var = np.sum((sales - mu) ** 2) / (len(sales) - 1)
print('variance', var)
std = var ** 0.5
print('std', std)
```

```python
# 結論
# numpy  -> 母體標準差：(sum((xi - x(average))** 2) / total)** 0.5
# pandas -> 樣本標準差：(sum((xi - x(average))** 2) / (total - 1))** 0.5
```

```python
# 變異係數
# std / mean <--numpy.std 視為母體，所以除以 N

import pandas as pd
import numpy as np

lemon = pd.read_csv('Lemonade.csv')
lemon.Temperature.values.std() / lemon.Temperature.values.mean()
```

#### 偏態係數

```python
# 偏態係數：指一組資料分配不對稱的程度
# SK＝ ((mu - median) / std ) ** 3

import pandas as pd
import numpy as np

lemon = pd.read_csv('Lemonade.csv')
X = lemon.Temperature.values

((np.mean(X) - np.median(X)) / np.std(X)) ** 3 # 接近0，表示對稱分布
```

#### 峰度係數

```python
# 峰度係數

import pandas as pd
import numpy as np

lemon = pd.read_csv('Lemonade.csv')
X = lemon.Temperature.values

((X - np.mean(X)) ** 4).sum() / (len(X) * np.std(X) ** 4) - 3
```
