```python
# https://zh.wikipedia.org/zh-tw/%E6%AD%A3%E6%80%81%E5%88%86%E5%B8%83#/media/File:Normal_Distribution_PDF.svg

import numpy as np

def f(mu, sigma, x):
    return(1 / (sigma * (2 * np.pi) ** 0.5) * np.exp(-1 / 2 * ((x - mu) / sigma) ** 2))

X = np.linspace(-5, 5, 100)
Y = f(mu = 0, sigma = 0.2 ** 0.5, x = X)
Y
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(mu, sigma, x):
    return(1 / (sigma * (2 * np.pi) ** 0.5) * np.exp(-1 / 2 * ((x - mu) / sigma) ** 2))

X = np.linspace(-5, 5, 100)

Y = f(mu = 0, sigma = 0.2 ** 0.5, x = X)
plt.plot(X, Y, color = 'blue')

Y = f(mu = 0, sigma = 1.0 ** 0.5, x = X)
plt.plot(X, Y, color = 'red')

Y = f(mu = 0, sigma = 5.0 ** 0.5, x = X)
plt.plot(X, Y, color = 'yellow')

Y = f(mu = -2, sigma = 0.5 ** 0.5, x = X)
plt.plot(X, Y, color = 'green')

plt.grid()
plt.show()
```

```python
import numpy as np

def f(a, b, x):
    
#     if a <= x <= b:
#         return(1 / (b - a))
#     else:
#         return(0)

    return(1 / (b - a) if a <= x <= b else 0)

f(1, 6, 7)
```

## 離散型資料分佈
#### 二項分佈(Binomial Distribution)

```python
# 階乘

import math

math.factorial(3)
```

```python
# 二項式係數

import math

def C(n, k):
    return(math.factorial(n) / (math.factorial(k) * math.factorial(n-k)))

C(3, 2)
```

```python
# https://zh.wikipedia.org/zh-tw/%E4%BA%8C%E9%A0%85%E5%BC%8F%E5%88%86%E5%B8%83#/media/File:Binomial_distribution_pmf.svg

import numpy as np

def f(k, n, p):
    return(C(n, k) * p ** k * (1 - p) ** (n - k))

f(k = 20, n = 40, p = 0.5)
```

```python
# 承上，畫出分佈圖

import matplotlib.pyplot as plt

X = np.arange(0, 41)
Y = [f(k = x, n = 40, p = 0.5) for x in X]
print(Y)

plt.scatter(X, Y, color = 'red')
plt.show()
```

```python
import matplotlib.pyplot as plt

X = np.arange(0, 41)
Y = [f(k = x, n = 40, p = 0.5) for x in X]
plt.scatter(X, Y, color = 'red')

X = np.arange(0, 20)
Y = [f(k = x, n = 20, p = 0.5) for x in X]
plt.scatter(X, Y, color = 'blue')

X = np.arange(0, 20)
Y = [f(k = x, n = 20, p = 0.7) for x in X]
plt.scatter(X, Y, color = 'green')

plt.show()
```

#### 卜瓦松分佈(Poisson Distribution)

```python
# https://zh.wikipedia.org/zh-tw/%E5%8D%9C%E7%93%A6%E6%9D%BE%E5%88%86%E5%B8%83#/media/File:Poisson_pmf.svg

import numpy as np
import math

def f(lamb, k):
    return((np.exp(-1 * lamb) * lamb ** k) / math.factorial(k))

f(10, 10)
```

```python
# 承上，畫出分佈圖

import matplotlib.pyplot as plt

X = range(20)
Y = [f(lamb = 10, k = x) for x in X]

plt.scatter(X, Y, color = 'aqua')
plt.plot(X, Y, color = 'gray')
plt.show()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt

def f(lamb, k):
    return((np.exp(-1 * lamb) * lamb ** k) / math.factorial(k))

X = range(0, 20)
Y = [f(lamb = 10, k = x) for x in X]
plt.scatter(X, Y, color = 'aqua')
plt.plot(X, Y, color = 'gray')

X = np.arange(0, 20)
Y = [f(lamb = 1, k = x) for x in X]
plt.scatter(X, Y, color = 'orange')
plt.plot(X, Y, color = 'gray')

X = np.arange(0, 20)
Y = [f(lamb = 4, k = x) for x in X]
plt.scatter(X, Y, color = 'purple')
plt.plot(X, Y, color = 'gray')

plt.show()
```

## HW：固定機率選取資料
系統跑100000次，於1-6這六個數值選取  
而這六個數值的機率為0.1, 0.2, 0.3, 0.1, 0.2, 0.1  
試著產生圖表

```python
# np.random.choice()：choice(a, size=None, replace=True, p=None)

result = np.random.choice(a = [1, 2, 3, 4, 5, 6], size = (100000,), replace = True, p = [0.1, 0.2, 0.3, 0.1, 0.2, 0.1])
result
```

```python
# 根據上述，加入np.bincount

# np.bincount()：bincount(x, /, weights=None, minlength=0)
# Count number of occurrences of each value in array of non-negative ints.

import numpy as np

result = np.random.choice(a = [1, 2, 3, 4, 5, 6], size = (100000,), replace = True, p = [0.1, 0.2, 0.3, 0.1, 0.2, 0.1])
np.bincount(result)[-6:]
```

```python
# 繪製圖表

import numpy as np
import matplotlib.pyplot as plt

data = [1, 2, 3, 4, 5, 6]
result = np.random.choice(a = data, size = (100000,), replace = True, p = [0.1, 0.2, 0.3, 0.1, 0.2, 0.1])
print(np.bincount(result)[-6:])
plt.bar(data, np.bincount(result)[-6:], color = 'blue')
plt.show()
```

```python
# 驗證機率是否為0.1, 0.2, 0.3, 0.1, 0.2, 0.1

import numpy as np
import matplotlib.pyplot as plt

data = [1, 2, 3, 4, 5, 6]
total = 100000
result = np.random.choice(a = data, size = (total,), replace = True, p = [0.1, 0.2, 0.3, 0.1, 0.2, 0.1])
print((np.bincount(result)[-6:]) / np.sum(np.bincount(result)[-6:]))
```
